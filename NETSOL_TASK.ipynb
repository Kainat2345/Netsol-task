```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#normal
mean=0
std_dev=1
size=1000
data=np.random.normal(mean,std_dev,size)
plt.hist(data,bins=30,density=True,alpha=0.7,color='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

```python
#uniform
low=0
high=10
data=np.random.uniform(low,high,1000)
plt.hist(data,bins=30,density=True,alpha=0.7,color='green')
plt.title('Uniform Distribution')
plt.show()
```

```python
#bionomial
n=10
p=0.5
data=np.random.binomial(n,p,1000)
plt.hist(data,bins=30,density=True,alpha=0.7,color='red')
plt.title('Binomial Distribution')
plt.show()
```

```python
#poisson
lam=3
data=np.random.poisson(lam,1000)
plt.hist(data,bins=30,density=True,alpha=0.7,color='orange')
plt.title('Poisson Distribution')
plt.show()
```

```python
#bernoulli
p=0.5
data=np.random.binomial(1,p,1000)
values,counts=np.unique(data,return_counts=True)
plt.bar(values,counts/1000,color=['purple','green'])
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Bernoulli Distribution')
```

```python
#exponential
scale=1.0
size=1000
data=np.random.exponential(scale,size)
plt.hist(data,bins=30,density=True,alpha=0.7,color='blue')
plt.title('Exponential Distribution')
plt.show()
```

```python
#chi_square
df=4
size=1000
data=np.random.chisquare(df,size)
plt.hist(data,bins=30,density=True,alpha=0.7,color='green')
plt.show()
```
